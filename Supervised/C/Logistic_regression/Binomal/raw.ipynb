```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([0.5, 1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1, 1])
```

```python
plt.scatter(X, y)
plt.xlabel('Experience')
plt.ylabel('Seleted (0 or 1)')
plt.title('Raw Dataset')
plt.show()
```

```python
W = 0.0 # Slope / weight
b = 0.0 # intercept / bias

learning_rate = 0.1
epochs = 1000
```

```python
losses = []
for epoch in range(epochs):
    z = W * X + b
    y_hat = 1 / (1 + np.exp(-z))
    if epoch % 200 == 0:
        plt.scatter(X,y)
        x_vals = np.linspace(0, 9, 100)
        y_vals = 1 / (1 + np.exp(-(W * x_vals + b)))
        plt.plot(x_vals, y_vals, color='red')
        plt.title(f'Sigmoid Curve at Epoch {epoch}')
        plt.show()
    loss = -np.mean(
        y * np.log(y_hat + 1e-9)+
        (1 - y) * np.log(1 - y_hat + 1e-9)
    )
    losses.append(loss)
    dw = np.mean((y_hat - y) * X)
    db = np.mean(y_hat - y)
    W = W - learning_rate * dw
    b = b - learning_rate * db
```

```python
plt.plot(losses)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Reduction Curve")
plt.show()
```

```python
plt.scatter(X, y)
x_vals = np.linspace(0, 9, 100)
y_vals = 1 / (1 + np.exp(-(W * x_vals + b)))
plt.plot(x_vals, y_vals, color='red')
plt.axhline(0.5, linestyle='--', color='gray')
plt.xlabel("Experience")
plt.ylabel("Probability")
plt.title("Final Logistic Regression Model")
plt.show()
```

```python
probabilities = 1 / (1 + np.exp(-(W * X + b)))
predictions = (probabilities >= 0.5).astype(int)

print("Probabilities:", probabilities)
print("Predictions:", predictions)
```
