```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Experience": [0.5, 1, 2, 3, 4, 5, 6, 7, 8],
    "Selected":   [0,   0, 0, 0, 1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
X = df[["Experience"]]   # MUST be 2D for sklearn
y = df["Selected"]
```

```python
X
```

```python
y
```

```python
plt.scatter(X, y)
plt.xlabel("Years of Experience")
plt.ylabel("Selected (0 = No, 1 = Yes)")
plt.title("Raw Data Distribution")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
x_values = np.linspace(0, 9, 100).reshape(-1, 1)
y_prob = model.predict_proba(x_values)[:, 1]
```

```python
plt.scatter(X, y)
plt.plot(x_values, y_prob, color="red")
plt.xlabel("Experience")
plt.ylabel("Probability of Selection")
plt.title("Logistic Regression Sigmoid Curve")
plt.show()
```

```python
plt.scatter(X, y)
plt.axhline(y=0.5, color='gray', linestyle='--')
plt.plot(x_values, y_prob, color="red")
plt.xlabel("Experience")
plt.ylabel("Probability")
plt.title("Decision Boundary at 0.5")
plt.show()
```

```python
predictions = model.predict(X)
print("Predicted classes:", predictions)
```

```python
print("Weight (Coefficient):", model.coef_)
print("Bias (Intercept):", model.intercept_)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy = accuracy_score(y, predictions)
cm = confusion_matrix(y, predictions)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", cm)
```
