```python
import numpy as np
```

```python
X = np.array([
    [1, 12],
    [2, 13],
    [3, 14],
    [4, 15],
    [5, 16]
])
```

```python
print(X)
```

```python
y = np.array([20, 25, 32, 40, 48])
print(y)
```

```python
X_b = np.c_[np.ones((X.shape[0], 1)), X]
print(X_b)
```

```python
theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
print(theta)
```
